```python
# for Keras
import tensorflow
```

```python
# for splitting keras Dataset, scikit learn has an easy way: sklearn.model_selection.train_test_split(...)
import sklearn.model_selection
```

```python
# For creating tensors easily
import numpy as np
```

```python
seed = 8225
batch = 256
n = 54000 // 256 + 1
```

```python
# MNIST
(kerasTrainX, kerasTrainY), (kerasTestX, kerasTestY) = tensorflow.keras.datasets.mnist.load_data(path="mnist.npz")
kerasTrainX, kerasValidX, kerasTrainY, kerasValidY = sklearn.model_selection.train_test_split(kerasTrainX, kerasTrainY, test_size=0.1, shuffle=True, random_state=seed)
```

```python
# https://stackoverflow.com/a/59568642/10945691
import subprocess as sp
import os

def get_gpu_memory():
    ACCEPTABLE_AVAILABLE_MEMORY = 1024
    COMMAND = "nvidia-smi --query-gpu=memory.free --format=csv"
    memory_free_info = sp.check_output(COMMAND.split()).decode('ascii').split('\n')[1:-1]
    memory_free_values = [int(x.split()[0]) for x in memory_free_info]
    return memory_free_values

# After building 1 model from Keras, this value will drop by 4-5GB, not sure why Keras allocates so much
get_gpu_memory()
```

```python
Glayers = [
    tensorflow.keras.layers.Dense(784, input_shape=(10,), activation='relu'),
    # tensorflow.keras.layers.Reshape((28, 28)),
]
G = tensorflow.keras.models.Sequential(Glayers)
G.compile(loss='mse', optimizer=tensorflow.keras.optimizers.Adam(learning_rate=1e-3))
```

```python
# # Will display a warning after 8 runs
# model = tensorflow.keras.models.Sequential()
# model.add(tensorflow.keras.layers.Conv2DTranspose(1, 4, (4, 4)))
# dummy_input = np.ones((32, 4, 2, 1))
# pred = model.predict(dummy_input)
# print(pred.shape)
```

```python
Dlayers = [
    tensorflow.keras.layers.Input(shape=(784,)),
    # tensorflow.keras.layers.Dense(1, input_shape=(28, 28), activation='sigmoid')
    tensorflow.keras.layers.Dense(1),
]
D = tensorflow.keras.models.Sequential(Dlayers)
D.compile(loss='mse', optimizer=tensorflow.keras.optimizers.Adam(learning_rate=1e-3))
```

```python
G.predict(np.zeros((batch, 10))).shape
```

```python
D.predict(np.zeros((batch,784))).shape
```

```python
D.predict(G.predict(np.zeros((batch, 10)))).shape
```

```python
GD = tensorflow.keras.models.Sequential(Glayers + Dlayers)
GD.compile(loss='mse', optimizer=tensorflow.keras.optimizers.Adam(learning_rate=1e-3))
```

```python
GD.predict(np.zeros((batch, 10))).shape
```

```python
datax = [np.array(kerasTrainX[batch * split:batch * (split + 1)]) for split in range(n)]
```

```python
for gepoch in range(1):
    for depoch in range(gepoch + 1): # as the generator gets better we train the discriminator for longer
        print('training discriminator')
        gx = G.predict(np.random.random((54000, 10)))
        gx = [np.array(gx[batch * split:batch * (split + 1)]) for split in range(n)]
        for i in range(n):
            u = len(gx[i])
            D.fit(gx[i], np.zeros((u, 1)), verbose= i % 22 == 0)
            D.fit(datax[i].reshape((u, 784)), np.ones((u, 1)), verbose= i % 22 == 0)
    print('training generator')
    for l in Dlayers:
        l.trainable = False
    GD.fit(np.random.random((54000, 10)), np.zeros((54000, 1)), verbose=True)
    for l in Dlayers:
        l.trainable = True
```
